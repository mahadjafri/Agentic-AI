`ollama ls` lists all the models you've downloaded  
`ollama rm <model_name>` deletes the specified model from your downloads

# Compare Different Gemini Models

````python
model_name_arr = [
            "gemini-2.5-pro","gemini-2.5-flash","gemini-2.5-flash-lite",
            "gemini-2.0-flash","gemini-2.0-flash-lite","gemini-1.5-flash-8b"]
for model in model_name_arr:
    model_name = model
    response = gemini.chat.completions.create(model=model_name, messages=messages)
    answer = response.choices[0].message.content
    competitors.append(model_name)
    answers.append(answer)
````

````python
# So where are we?
print(competitors)
print(answers)
````

````python
# It's nice to know how to use "zip"
for competitor, answer in zip(competitors, answers):
    print(f"Competitor: {competitor}\n\n{answer}")
````

````python
# Let's bring this together - note the use of "enumerate"

together = ""
for index, answer in enumerate(answers):
    together += f"# Response from competitor {index+1}\n\n"
    together += answer + "\n\n"
````

````python
print(together)
````

````python
judge = f"""You are judging a competition between {len(competitors)} competitors.
Each model has been given this question:

{question}

Your job is to evaluate each response for clarity and strength of argument, and rank them in order of best to worst.
Respond with JSON, and only JSON, with the following format:
{{"results": [{{"competitor": 1, "score": 8}}, {{"competitor": 2, "score": 6}}, {{"competitor": 3, "score": 7}}]}}

Here are the responses from each competitor:

{together}

Now respond with the JSON with a score out of 10 for the competitors, nothing else. Do not include markdown formatting or code blocks."""
````

````python
print(judge)
````

````python
judge_messages = [{"role": "user", "content": judge}]
judge_messages
````

````python
# Judgement time!
response = gemini.chat.completions.create(
    model="gemini-2.5-pro",
    messages=judge_messages,
)
results = response.choices[0].message.content
print(results)
````

````python
# Parse the JSON results and create ranking
results_dict = json.loads(results.strip('```json\n').strip('\n```'))
ranks = results_dict["results"]

# Sort by score (highest to lowest)
sorted_ranks = sorted(ranks, key=lambda x: x["score"], reverse=True)

print("=== AI Model Competition Ranking ===")
print(f"Question: {question}\n")

for index, result in enumerate(sorted_ranks):
    competitor_index = result["competitor"] - 1  # Convert to 0-based index
    competitor_name = competitors[competitor_index]
    score = result["score"]
    print(f"Rank {index+1}: {competitor_name} (Score: {score}/10)")
````
